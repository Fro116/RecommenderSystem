```julia
merge_edges::Dict{String,Vector{edgetype}} =
    Dict("anime" => [], "manga" => [(("mal", "27379"), ("mal", "3335"))])
# merge animeplanet manga that are split into parts
df = CSV.read(
    "../../../data/media/sources/animeplanet.manga_relations.csv",
    DataFrames.DataFrame,
)
df = filter(x -> x.source_media == "manga" && x.target_media == "manga", df)
for (s, t) in zip(df.source_id, df.target_id)
    prefix = s * "-part-"
    if startswith(t, prefix) && all(isdigit, t[length(prefix)+1:end])
        push!(merge_edges["manga"], (("animeplanet", s), ("animeplanet", t)))
    end
end
```

```julia
for m in ["manga", "anime"]
    edges = vcat(invalid_edges[m], valid_edges[m], overlapping_edges[m], merge_edges[m])
    @assert length(edges) == length(Set(edges)) "duplicate $m edges"
end
for m in ["manga", "anime"]
    num_edges = sum(length.([invalid_edges[m], valid_edges[m], overlapping_edges[m]]))
    @info "using $num_edges $m manual overrides"
end
for m in ["manga", "anime"]
    edges = [e[1] for e in valid_edges[m]]
    @assert length(edges) == length(Set(edges)) "$m edge is mismatched"
end
```

```julia
@memoize function get_valid_ids(medium, source)
    df = CSV.read(
        "../../../data/media/sources/$source.$medium.csv",
        DataFrames.DataFrame;
        ntasks = 1,
        missingstring = nothing,
        types = String,
    )
    Set(df.uid)
end;
```

```julia
function save(name::String, edges::Dict)
    outdir = "../../../data/media/manual/$name"
    mkpath(outdir)
    sources = ["mal", "anilist", "kitsu", "animeplanet"]
    for medium in ["anime", "manga"]
        for i = 1:length(sources)
            for j = i:length(sources)
                sj = get_valid_ids(medium, sources[j])
                si = get_valid_ids(medium, sources[i])
                open("$outdir/$medium.$(sources[j]).$(sources[i]).csv", "w") do f
                    write(f, "$(sources[j]),$(sources[i])\n")
                    for ((a, b), (c, d)) in edges[medium]
                        if a == sources[j] && c == sources[i]
                            if b in sj && d in si
                                write(f, "$b,$d\n")
                            else
                                @info (
                                    "skipping edge $(((a, b), (c, d))) with unknown vertex " *
                                    "$(b in sj) $(d in si)"
                                )
                            end
                        end
                    end
                end
            end
        end
    end
end;
```
